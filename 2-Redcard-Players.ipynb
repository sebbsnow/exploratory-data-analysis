# DEFINIR LES FONCTIONS

```python
# Redefinir la fonction "save_subgroup"
def save_subgroup(dataframe, g_index, subgroup_name, prefix='raw_'):
    save_subgroup_filename = "".join([prefix, subgroup_name, ".csv.gz"])
    dataframe.to_csv(save_subgroup_filename, compression='gzip', encoding='UTF-8')
    test_df = pd.read_csv(save_subgroup_filename, compression='gzip', index_col=g_index, encoding='UTF-8')
    # Test that we recover what we send in
    if dataframe.equals(test_df):
        print("Test-passed: we recover the equivalent subgroup dataframe.")
    else:
        print("Warning -- equivalence test!!! Double-check.")
```

```python
# Redefinir la fonction "load_subgroup"
def load_subgroup(filename, index_col=[0]):
    return pd.read_csv(filename, compression='gzip', index_col=index_col)
```

```python
# length of the entire players table
print("All players:", len(players))

# length where the rater1 value is a null value
print("rater1 nulls:", len(players[players.rater1.isnull()]))

# Gives a count where rater2 is a null value
print("rater2 nulls:", len(players[players.rater2.isnull()]))

# Gives a count where both values are both null
print("Both nulls:", len(players[(players.rater1.isnull()) & (players.rater2.isnull())]))
```

Nous avons 2 types d'évaluation (rater1 et rater2)

Est ce que les 2 rating sont correlés ?

Est ce qu'ils sont assez proche pour que nous puissions les combiner ?

```python
# Correlation entre rater1 et rater2
fig, ax = plt.subplots(figsize=(12, 8))
sns.heatmap(pd.crosstab(players.rater1, players.rater2), cmap='Blues', annot=True, fmt='d', ax=ax)
ax.set_title("Correlation between Rater 1 and Rater 2\n")
fig.tight_layout()

# La diagonale est quand les 2 rating sont les mêmes (correlés)
# et plus on s'éloigne de la diagonale et plus les ratings sont différents (pas correlés)
```

Les valeurs ne sont pas éloignées de la diagonale donc cela veut dire que les 2 raters sont globalement du même avis et que les 2 rating sont plutot bien correlés

### Create useful new columns

This looks correlated enough to me -- let's combine the rater's skintone ratings into a new column that is the average rating.


```python
# modifying dataframe en créant une colonne qui est la moyenne des colonnes "rater1" et "rater2"
players['skintone'] = players[['rater1', 'rater2']].mean(axis=1)

# Très utile en EDA => Créer une nouvelle colonne qui sera utile pour moi

# Afficher les premières lignes
players.head()
```

# Visualize distributions of univariate features

What is the skintone distribution?

```python
# Nombre de players en fonction du skintone rating
# By default, "distplot" will draw a histogram and fit a kernel density estimate (KDE)
# on peut ne pas afficher le KDE en mettant la valeur False

# Lien pour les graphiques
# https://seaborn.pydata.org/tutorial/aesthetics.html

sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(20, 10)
sns.distplot(players.skintone, kde=False, bins=int(100/5));
```

## Positions

Might the player's position correlate with the baseline susceptibility to redcards? Likely that a defender would have a higher rate than a keeper, for example.

```python
# Afficher le nombre joueurs en fonction des positions
# Afficher par ordre decroissant par defaut / ascending=True permet d'afficher par orre croissant
players.position.value_counts(dropna=False)
```

```python
# Afficher le graphique du nombre de joueurs en fonction de la position
MIDSIZE = (12, 8)
fig, ax = plt.subplots(figsize=MIDSIZE)
fig.set_size_inches(20, 10)

# Définir les éléments a afficher
players.position.value_counts(dropna=False, ascending=True).plot(kind='barh', ax=ax)

# Définir les labels
ax.set_ylabel("Position")
ax.set_xlabel("Counts")
fig.tight_layout()
```
